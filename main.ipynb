```python
import GeneticAlgorithm.GeneticAlgorithm as ga
from tools.City import representation
from tools.functions import *
```

```python
# Set model parameters
popSize = 10
crossoverRate = 0.9
mutationRate = 0.9
num_iter = 5*int(1e3)
```

```python
# Load cities and compute the distance matrix
cities = load('p01_sxy.txt', ';')
showCities(cities)
EDM = computeEDM(cities)

# Permutation representation of our phenotype (cities)
chromosome = representation(cities)
```

```python
# Instanciate our model
genAlg = ga.GeneticAlgorithm(chromosome, popSize, crossoverRate, mutationRate, num_iter, show = True)

# Set optimal value
genAlg.optimal_value = 291 # 33523 

# Set termination condition
genAlg.epsilon = 5
```

```python
# Run our Genetic Algorithm model
fittest, bestFitness = genAlg.run(EDM)
```

```python
# Show solution
showPath(fittest)
```
